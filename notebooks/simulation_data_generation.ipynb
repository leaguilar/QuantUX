```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
```

```python
# 3 latent factors
n = 300
F1 = np.random.normal(size=n)
F2 = np.random.normal(size=n)
F3 = np.random.normal(size=n)
```

```python
# observed variables (add some noise)
data = pd.DataFrame({
    'x1': 0.8*F1 + np.random.normal(0, 0.4, n),
    'x2': 0.7*F1 + np.random.normal(0, 0.4, n),
    'x3': 0.9*F1 + np.random.normal(0, 0.4, n),
    'x4': 0.8*F2 + np.random.normal(0, 0.4, n),
    'x5': 0.7*F2 + np.random.normal(0, 0.4, n),
    'x6': 0.9*F2 + np.random.normal(0, 0.4, n),
    'x7': 0.8*F3 + np.random.normal(0, 0.4, n),
    'x8': 0.7*F3 + np.random.normal(0, 0.4, n),
    'x9': 0.9*F3 + np.random.normal(0, 0.4, n),
})
```
